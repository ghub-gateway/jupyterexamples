```python
import plotly
import plotly.graph_objs as go
import numpy as np

# The old way to use plotly in Jupyter is to use iplot
# either online (if you have an account) or offline
from plotly.offline import iplot
plotly.offline.init_notebook_mode(connected=True)

N = 500
random_x = np.linspace(0, 1, N)
random_y = np.random.randn(N)

# Create a trace
trace = go.Scatter(
    x = random_x,
    y = random_y
)

data = [trace]

iplot(data)
```

```python
# The new way to use plotly is to use FigureWidget()
# FigureWidgets work like other Jupyter widgets.

data = [go.Bar(x=['A', 'B', 'C'],
            y=[2,4,6])]

go.FigureWidget(data)
```

```python
s = np.linspace(0, 2 * np.pi, 240)
t = np.linspace(0, np.pi, 240)
tGrid, sGrid = np.meshgrid(s, t)

r = 2 + np.sin(7 * sGrid + 5 * tGrid)  # r = 2 + sin(7s+5t)
x = r * np.cos(sGrid) * np.sin(tGrid)  # x = r*cos(s)*sin(t)
y = r * np.sin(sGrid) * np.sin(tGrid)  # y = r*sin(s)*sin(t)
z = r * np.cos(tGrid)                  # z = r*cos(t)

f2 = go.FigureWidget()
f2.add_surface(x=x, y=y, z=z)
f2
```

```python
layout = go.Layout(
    title='Parametric Plot',
    scene=go.Scene(
        xaxis=go.XAxis(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        yaxis=go.YAxis(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        zaxis=go.ZAxis(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        )
    )
)
f2.layout = layout
```

```python
A=1
B=0
x = np.linspace(0,10,200)
y = np.sin(A*x+B)

fig = go.FigureWidget()
sline = fig.add_scatter(line={'color': 'red', 'width': 3},
            name="A=%s, B=%s" % (A, B),
            x=x,
            y=y
        )
fig
```

```python
def update_plot(A, B, color):
    y = np.sin(A*x+B)
    with fig.batch_update():
        sline.line.color = color
        sline.y = y
```

```python
from ipywidgets import interact
interact(update_plot, A=(0,10), B=(0,10), color=['red', 'green', 'blue']);
```
