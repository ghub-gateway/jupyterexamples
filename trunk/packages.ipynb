# Currently Installed Packages

```python
import pandas as pd
import json
import urllib.request
import qgrid
from ipywidgets import HTML
```

```python
# get information from pypi database
def get_info(name, ver):
    try:
        with urllib.request.urlopen('https://pypi.python.org/pypi/%s/%s/json' % (name, ver)) as response:
           d = json.loads(response.read().decode('utf-8'))
    except:
        return None, None, None
    d = d['info']
    return d['description'], d['home_page'], d['summary']
```

```python
# get a list of the installed packages
def get_package_list():
    !pip list --format json > packages.json

    with open('packages.json') as json_file:  
        pack_list = json.load(json_file)

    data = []
    for item in pack_list:
        name, ver = item['name'], item['version']
        desc, url, summary = get_info(name, ver)
        if url is not None:
            url = '<a href="{}" target="_blank">{}</a>'.format(url, url)
        data.append({'Name': name, 'Version': ver, 'URL': url, 'Summary': summary})
    return data
```

```python
h = HTML(value="<h2>Building package list ...</h2>")
display(h)
data = get_package_list()
df = pd.DataFrame.from_dict(data)
df = df[['Name', 'Version', 'Summary', 'URL']].set_index('Name')
qw = qgrid.show_grid(df, grid_options={'editable': False})
h.value=''
display(qw)
```
